# Complementing a Strand of DNA
http://rosalind.info/problems/revc/

### A solution

```python
def attempt1(dna):
    complement = {
        'A': 'T',
        'C': 'G',
        'G': 'C',
        'T': 'A'
    }
    result = ''
    for base in dna[::-1]:
        result += complement[base]

    return result
```
