```python
def translate(mRNA):
    translated = []
    for i in range(0, len(mRNA), 3):
        m = mRNA[i:i+3]
        translated.append(genetic_code[m])

    return ''.join(translated)
```

### Another solution

```python
def translate2(input_string):
    amino_string = ''
    for i, char in enumerate(input_string):
        if i % 3 == 0:
            codon = input_string[i:i+3]
            amino = genetic_code[codon]
            if amino == 'STOP':
                break
            amino_string += amino
```

```python
%timeit translated = translate2(mRNA)
```
